```python
import sentence_transformers
import buscador
import otimizador
```

# Downloading a pretrained Ulysses SBERT model

We'll used Ulysses Fetcher to download a pretrained Ulysses SBERT model. Check [Ulysses Fetcher documentation](https://github.com/ulysses-camara/ulysses-fetcher) for more information regarding this step. This step can be skipped if yuou already have an pretrained SBERT model at hand.

```python
buscador.download_model(
    task_name="sentence_similarity",
    model_name="distil_sbert_br_ctimproved_12_epochs_v1",
    output_dir="./pretrained_models",
)
```

```python
sbert = sentence_transformers.SentenceTransformer(
    "./pretrained_models/distil_sbert_br_ctimproved_12_epochs_v1/",
    device="cpu"
)
```

# Quantizing a SBERT (Sentence BERT) model

```python
quantized_model_paths = otimizador.sbert.quantize_as_onnx(sbert, task_name="quantization_test")
```

```python
quantized_model_paths
```

# Loading a Quantized SBERT model (in ONNX format)

```python
quantized_model = otimizador.sbert.ONNXSBERT(
    uri_model=quantized_model_paths.output_uri,
    uri_tokenizer=sbert.tokenizer.name_or_path,
)
```

```python
quantized_model(["Teste de inferência."])
```
